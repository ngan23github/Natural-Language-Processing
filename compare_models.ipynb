```python
import matplotlib.pyplot as plt
# ====== So sánh dữ liệu BLEU ======
models = ["Base", "Attention", "Attention + Beam"]
scores = [0.1467338631792324, 0.2478587343519596, 0.26460954539669834]

# ====== Vẽ biểu đồ ======
plt.figure(figsize=(8, 6))
plt.bar(models, scores)

# Nhãn & tiêu đề
plt.ylabel("BLEU Score")
plt.xlabel("Các mô hình")
plt.title("So sánh BLEU Score giữa các mô hình")
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Hiển thị giá trị BLEU trên đầu cột
for i, score in enumerate(scores):
    plt.text(i, score + 0.005, f"{score:.4f}", ha='center', fontsize=10)

# Lưu file (tùy chọn)
plt.tight_layout()
plt.savefig("bleu_scores.png", dpi=300)

# Hiển thị
plt.show()
```
